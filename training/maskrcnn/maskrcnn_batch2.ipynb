```python
# Mount Google Drive BEFORE training
import subprocess
import threading
import time
import sys

def show_progress():
    """Show visual progress while waiting for training to start"""
    spinner = ['⠋', '⠙', '⠹', '⠸', '⠼', '⠴', '⠦', '⠧', '⠇', '⠏']
    idx = 0
    start = time.time()

    while not started and not stopped:
        elapsed = int(time.time() - start)
        mins = elapsed // 60
        secs = elapsed % 60
        print(f'\r{spinner[idx % len(spinner)]} Initializing training... {mins}m {secs}s | Downloading weights & loading data...', end='', flush=True)
        idx += 1
        time.sleep(0.1)

    if not stopped:
        print('\r✅ Training started! Showing live output below...                    ')

# Start progress indicator
started = False
stopped = False
progress_thread = threading.Thread(target=show_progress, daemon=True)
progress_thread.start()

# Run training and capture output
proc = subprocess.Popen(
    ['conda', 'run', '--no-capture-output', '-n', 'openmmlab',
     'python', 'tools/train.py',
     'configs/custom/maskrcnn_car_parts_batch2.py',  # 🔥 FIXED filename
     '--work-dir', '/content/work_dirs/maskrcnn_batch2'],  # 🔥 FIXED work_dir
    cwd='/content/mmdetection',
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    universal_newlines=True,
    bufsize=1
)

# Stream output line by line
try:
    for line in proc.stdout:
        if not started:
            started = True  # Stop spinner when first output appears
            time.sleep(0.2)  # Let spinner finish
        print(line, end='', flush=True)
except KeyboardInterrupt:
    stopped = True
    proc.terminate()
    print("\n⚠️ Training interrupted by user")

proc.wait()

if proc.returncode != 0:
    print(f"\n❌ Training failed with exit code {proc.returncode}")
else:
    print("\n✅ Training completed successfully!")
```

```python
import shutil
from pathlib import Path
from google.colab import files, drive
import zipfile
from tqdm import tqdm

WORK_DIR = Path("/content/work_dirs/maskrcnn_batch2")

if WORK_DIR.exists():
    # Calculate total size first
    print("📊 Calculating folder size...")
    total_size = sum(f.stat().st_size for f in WORK_DIR.rglob("*") if f.is_file())
    total_mb = total_size / (1024*1024)
    file_count = sum(1 for f in WORK_DIR.rglob("*") if f.is_file())

    print(f"   Total: {file_count} files, {total_mb:.2f} MB")

    # Create zip with progress bar
    print("\n📦 Creating zip archive...")
    zip_path = '/content/maskrcnn_batch2_training.zip'

    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        files_list = list(WORK_DIR.rglob("*"))
        for file in tqdm(files_list, desc="Zipping files", unit="file"):
            if file.is_file():
                arcname = file.relative_to(WORK_DIR.parent)
                zipf.write(file, arcname)

    zip_size = Path(zip_path).stat().st_size / (1024*1024)
    print(f"✅ Zip created: {zip_size:.2f} MB")

    # 1. Save to Google Drive
    print("\n☁️ Mounting Google Drive...")
    drive.mount('/content/drive', force_remount=False)

    print("📤 Copying to Google Drive...")
    DRIVE_PATH = Path("/content/drive/MyDrive")

    # Copy with progress
    chunk_size = 1024 * 1024  # 1MB chunks
    with open(zip_path, 'rb') as src:
        with open(DRIVE_PATH / 'maskrcnn_batch2_training.zip', 'wb') as dst:
            total_chunks = int(zip_size)
            with tqdm(total=total_chunks, desc="Uploading", unit="MB") as pbar:
                while True:
                    chunk = src.read(chunk_size)
                    if not chunk:
                        break
                    dst.write(chunk)
                    pbar.update(1)

    print(f"✅ Saved to Drive: {DRIVE_PATH}/maskrcnn_batch2_training.zip")

    # 2. Download to PC
    print("\n💾 Downloading to your PC...")
    files.download(zip_path)
    print("✅ Download started! Check your browser downloads.")

    # Summary
    print(f"\n📊 Summary:")
    print(f"   Files: {file_count}")
    print(f"   Original: {total_mb:.2f} MB")
    print(f"   Compressed: {zip_size:.2f} MB")
    print(f"   Saved {(1 - zip_size/total_mb)*100:.1f}% space")

else:
    print("❌ Training folder doesn't exist yet!")
```
