## Create HTC Training Configuration

⚠️ **SEARCH REQUIRED**: You need to find the following from MMDetection documentation:

1. **HTC base config path** - Search: "htc base config mmdetection"
   - Likely path: `../_base_/models/htc_r50_fpn.py` or `../_base_/models/htc-r50-fpn.py`
   - Update the `_base_` list below with correct path

2. **HTC pretrained weights URL** - Search: "htc pretrained weights openmmlab download"
   - Pattern: `https://download.openmmlab.com/mmdetection/v2.0/htc/htc_...`
   - Need 3x COCO trained version (similar to Mask R-CNN)
   - Update `load_from` variable below

3. **HTC model structure** - Search: "htc cascade_roi_head mmdetection config"
   - HTC uses `cascade_roi_head` with 3 stages
   - Each stage needs `num_classes` specified
   - Verify the model dict structure below matches MMDetection's format

```python
from pathlib import Path

WORK_DIR = Path("/content/work_dirs/htc_batch2")
WORK_DIR.mkdir(parents=True, exist_ok=True)

CONFIG_DIR = Path("/content/mmdetection/configs/custom")
CONFIG_DIR.mkdir(parents=True, exist_ok=True)

config_text = f"""
_base_ = '../htc/htc-without-semantic_r50_fpn_1x_coco.py'

# Override num_classes for your dataset
model = dict(
    roi_head=dict(
        bbox_head=[
            dict(type='Shared2FCBBoxHead', num_classes={NUM_CLASSES}),
            dict(type='Shared2FCBBoxHead', num_classes={NUM_CLASSES}),
            dict(type='Shared2FCBBoxHead', num_classes={NUM_CLASSES})
        ],
        mask_head=[
            dict(type='HTCMaskHead', num_classes={NUM_CLASSES}),
            dict(type='HTCMaskHead', num_classes={NUM_CLASSES}),
            dict(type='HTCMaskHead', num_classes={NUM_CLASSES})
        ]
        # ← DON'T include semantic_head here!
    )
)

# Dataset settings
dataset_type = 'CocoDataset'
data_root = '/content/mmdetection/data/car_parts/'

classes = {tuple(CLASS_NAMES)}

# Override train pipeline - remove semantic segmentation
train_pipeline = [
    dict(backend_args=None, type='LoadImageFromFile'),
    dict(type='LoadAnnotations', with_bbox=True, with_mask=True),  # ← Remove with_seg=True
    dict(keep_ratio=True, scale=(1333, 800), type='Resize'),
    dict(prob=0.5, type='RandomFlip'),
    dict(type='PackDetInputs'),
]

train_dataloader = dict(
    dataset=dict(
        type=dataset_type,
        data_root=data_root,
        metainfo=dict(classes=classes),
        ann_file='train/_annotations.coco.json',
        data_prefix=dict(img='train/'),
        pipeline=train_pipeline  # ← Use the overridden pipeline
    )
)

val_dataloader = dict(
    dataset=dict(
        type=dataset_type,
        data_root=data_root,
        metainfo=dict(classes=classes),
        ann_file='valid/_annotations.coco.json',
        data_prefix=dict(img='valid/')
    )
)

test_dataloader = val_dataloader

val_evaluator = dict(
    type='CocoMetric',
    ann_file=data_root + 'valid/_annotations.coco.json',
    metric=['bbox', 'segm']
)
test_evaluator = val_evaluator

# Override training schedule to 24 epochs
train_cfg = dict(type='EpochBasedTrainLoop', max_epochs={EPOCHS}, val_interval=3)

param_scheduler = [
    dict(type='LinearLR', start_factor=0.001, by_epoch=False, begin=0, end=500),
    dict(type='MultiStepLR', begin=0, end={EPOCHS}, by_epoch=True, milestones=[24, 32], gamma=0.1)
]

load_from = 'https://download.openmmlab.com/mmdetection/v2.0/htc/htc_r50_fpn_1x_coco/htc_r50_fpn_1x_coco_20200317-7332cf16.pth'

work_dir = '{WORK_DIR.absolute()}'

default_hooks = dict(
    checkpoint=dict(
        type='CheckpointHook',
        interval=3,
        save_best='coco/segm_mAP',
        rule='greater'
    )
)
"""

config_path = CONFIG_DIR / "htc_car_parts_batch2.py"
with open(config_path, 'w') as f:
    f.write(config_text)

print(f"✅ Config saved to: {config_path}")
print(f"📋 Training Configuration:")
print(f"   Base: htc_r50_fpn_1x_coco.py")
print(f"   Epochs: {EPOCHS}")
print(f"   Classes: {NUM_CLASSES}")
```

## Run Training

⚠️ **IMPORTANT**: Make sure you've searched and updated the config values in the previous cell before running this!

```python
# Mount Google Drive BEFORE training
from google.colab import drive
print("☁️ Mounting Google Drive for checkpoint backups...")
drive.mount('/content/drive', force_remount=False)
print("✅ Drive mounted!\n")

import subprocess
import threading
import time
import sys

def show_progress():
    """Show visual progress while waiting for training to start"""
    spinner = ['⠋', '⠙', '⠹', '⠸', '⠼', '⠴', '⠦', '⠧', '⠇', '⠏']
    idx = 0
    start = time.time()

    while not started and not stopped:
        elapsed = int(time.time() - start)
        mins = elapsed // 60
        secs = elapsed % 60
        print(f'\r{spinner[idx % len(spinner)]} Initializing training... {mins}m {secs}s | Downloading weights & loading data...', end='', flush=True)
        idx += 1
        time.sleep(0.1)

    if not stopped:
        print('\r✅ Training started! Showing live output below...                    ')

# Start progress indicator
started = False
stopped = False
progress_thread = threading.Thread(target=show_progress, daemon=True)
progress_thread.start()

# 🔥 FIXED: Use correct config filename
# Run training and capture output
proc = subprocess.Popen(
    ['conda', 'run', '--no-capture-output', '-n', 'openmmlab',
     'python', 'tools/train.py',
     'configs/custom/htc_car_parts_batch2.py',  # 🔥 FIXED filename
     '--work-dir', '/content/work_dirs/htc_batch2'],  # 🔥 FIXED work_dir
    cwd='/content/mmdetection',
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    universal_newlines=True,
    bufsize=1
)

# Stream output line by line
try:
    for line in proc.stdout:
        if not started:
            started = True  # Stop spinner when first output appears
            time.sleep(0.2)  # Let spinner finish
        print(line, end='', flush=True)
except KeyboardInterrupt:
    stopped = True
    proc.terminate()
    print("\n⚠️ Training interrupted by user")

proc.wait()

if proc.returncode != 0:
    print(f"\n❌ Training failed with exit code {proc.returncode}")
else:
    print("\n✅ Training completed successfully!")
```

## Save Training Results

```python
import shutil
from pathlib import Path
from google.colab import files, drive
import zipfile
from tqdm import tqdm

WORK_DIR = Path("/content/work_dirs/htc_batch2")

if WORK_DIR.exists():
    # Calculate total size first
    print("📊 Calculating folder size...")
    total_size = sum(f.stat().st_size for f in WORK_DIR.rglob("*") if f.is_file())
    total_mb = total_size / (1024*1024)
    file_count = sum(1 for f in WORK_DIR.rglob("*") if f.is_file())

    print(f"   Total: {file_count} files, {total_mb:.2f} MB")

    # Create zip with progress bar
    print("\n📦 Creating zip archive...")
    zip_path = '/content/htc_batch2_training.zip'

    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        files_list = list(WORK_DIR.rglob("*"))
        for file in tqdm(files_list, desc="Zipping files", unit="file"):
            if file.is_file():
                arcname = file.relative_to(WORK_DIR.parent)
                zipf.write(file, arcname)

    zip_size = Path(zip_path).stat().st_size / (1024*1024)
    print(f"✅ Zip created: {zip_size:.2f} MB")

    # 1. Save to Google Drive
    print("\n☁️ Mounting Google Drive...")
    drive.mount('/content/drive', force_remount=False)

    print("📤 Copying to Google Drive...")
    DRIVE_PATH = Path("/content/drive/MyDrive")
    drive_file = DRIVE_PATH / 'htc_batch2_training.zip'

    # Reliable copy
    shutil.copy2(zip_path, drive_file)

    # CRITICAL: Force Drive to flush writes
    print("⏳ Syncing to Drive (this may take a minute)...")
    drive.flush_and_unmount()
    drive.mount('/content/drive')

    # Verify the upload
    if drive_file.exists():
        drive_size = drive_file.stat().st_size / (1024*1024)
        local_size = Path(zip_path).stat().st_size / (1024*1024)

        if abs(drive_size - local_size) < 1:  # Within 1MB tolerance
            print(f"✅ Verified on Drive: {drive_size:.2f} MB")
        else:
            print(f"❌ Size mismatch! Local: {local_size:.2f} MB, Drive: {drive_size:.2f} MB")
    else:
        print("❌ File not found on Drive!")

    # 2. Download to PC
    print("\n💾 Downloading to your PC...")
    files.download(zip_path)
    print("✅ Download started! Check your browser downloads.")

    # Summary
    print(f"\n📊 Summary:")
    print(f"   Files: {file_count}")
    print(f"   Original: {total_mb:.2f} MB")
    print(f"   Compressed: {zip_size:.2f} MB")
    print(f"   Saved {(1 - zip_size/total_mb)*100:.1f}% space")

else:
    print("❌ Training folder doesn't exist yet!")
```
