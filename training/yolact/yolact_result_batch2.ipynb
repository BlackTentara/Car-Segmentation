# YOLACT Training Results Analysis - Batch 2

This notebook analyzes the YOLACT batch 2 training results and extracts validation metrics.

```python
# Import libraries
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
# Load training data
WORK_DIR = Path('yolact_batch2')
training_dirs = [d for d in WORK_DIR.glob('202*') if d.is_dir()]
latest_dir = sorted(training_dirs)[-1]
print(f"Using training directory: {latest_dir}")

# Get log file
log_file = list(latest_dir.glob('*.log'))[0]
print(f"Log file: {log_file}")

# Create results directory
results_dir = Path('yolact_batch2_results')
results_dir.mkdir(exist_ok=True)
```

```python
# Extract final validation results from log
with open(log_file, 'r') as f:
    log_content = f.read()
    log_lines = log_content.split('\n')

# Extract bbox metrics
bbox_line = [l for l in log_lines if 'bbox_mAP_copypaste' in l][-1]
bbox_values = bbox_line.split('bbox_mAP_copypaste:')[1].strip().split()
bbox_mAP, bbox_mAP_50, bbox_mAP_75, bbox_mAP_s, bbox_mAP_m, bbox_mAP_l = map(float, bbox_values)

print(f"BBox Metrics (Final Validation):")
print(f"  mAP: {bbox_mAP*100:.1f}%")
print(f"  mAP@50: {bbox_mAP_50*100:.1f}%")
print(f"  mAP@75: {bbox_mAP_75*100:.1f}%")

# Extract segmentation metrics
segm_line = [l for l in log_lines if 'segm_mAP_copypaste' in l][-1]
segm_values = segm_line.split('segm_mAP_copypaste:')[1].strip().split()
segm_mAP, segm_mAP_50, segm_mAP_75, segm_mAP_s, segm_mAP_m, segm_mAP_l = map(float, segm_values)

print(f"\nSegmentation Metrics (Final Validation):")
print(f"  mAP: {segm_mAP*100:.1f}%")
print(f"  mAP@50: {segm_mAP_50*100:.1f}%")
print(f"  mAP@75: {segm_mAP_75*100:.1f}%")
```

```python
# Create summary table
summary = pd.DataFrame({
    'Method': ['YOLACT Batch 2'],
    'bbox_mAP': [f"{bbox_mAP*100:.1f}"],
    'bbox_mAP_50': [f"{bbox_mAP_50*100:.1f}"],
    'bbox_mAP_75': [f"{bbox_mAP_75*100:.1f}"],
    'segm_mAP': [f"{segm_mAP*100:.1f}"],
    'segm_mAP_50': [f"{segm_mAP_50*100:.1f}"],
    'segm_mAP_75': [f"{segm_mAP_75*100:.1f}"]
})

print("\nSummary Table:")
print(summary)

# Save to CSV
summary.to_csv(results_dir / 'yolact_batch2_summary.csv', index=False)
print(f"\nSaved summary to {results_dir / 'yolact_batch2_summary.csv'}")
```

```python
# Create detailed bbox metrics table
bbox_metrics = pd.DataFrame({
    'Metric': ['mAP', 'mAP@50', 'mAP@75', 'mAP_s', 'mAP_m', 'mAP_l'],
    'Value': [f"{bbox_mAP*100:.1f}", f"{bbox_mAP_50*100:.1f}", f"{bbox_mAP_75*100:.1f}",
              f"{bbox_mAP_s*100:.1f}", f"{bbox_mAP_m*100:.1f}", f"{bbox_mAP_l*100:.1f}"]
})

print("\nBBox Metrics:")
print(bbox_metrics)

bbox_metrics.to_csv(results_dir / 'yolact_batch2_bbox.csv', index=False)
print(f"Saved bbox metrics to {results_dir / 'yolact_batch2_bbox.csv'}")
```

```python
# Create detailed segmentation metrics table
segm_metrics = pd.DataFrame({
    'Metric': ['mAP', 'mAP@50', 'mAP@75', 'mAP_s', 'mAP_m', 'mAP_l'],
    'Value': [f"{segm_mAP*100:.1f}", f"{segm_mAP_50*100:.1f}", f"{segm_mAP_75*100:.1f}",
              f"{segm_mAP_s*100:.1f}", f"{segm_mAP_m*100:.1f}", f"{segm_mAP_l*100:.1f}"]
})

print("\nSegmentation Metrics:")
print(segm_metrics)

segm_metrics.to_csv(results_dir / 'yolact_batch2_segm.csv', index=False)
print(f"Saved segmentation metrics to {results_dir / 'yolact_batch2_segm.csv'}")
```

## Summary

YOLACT batch 2 validation results have been extracted and saved to CSV files in the `yolact_batch2_results` directory:
- `yolact_batch2_summary.csv`: Overall metrics comparison
- `yolact_batch2_bbox.csv`: Detailed bounding box metrics
- `yolact_batch2_segm.csv`: Detailed segmentation metrics
